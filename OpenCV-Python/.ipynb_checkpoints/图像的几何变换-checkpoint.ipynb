# 图像的几何变换
主要用到这两个函数
```python
cv2.warpAffine()# 接收的参数是2*3的变换矩阵
cv2.warpPerspective()# 接收的参数是3*3的变换矩阵
```

````python
import cv2
import numpy as np
from matplotlib import pyplot as plt
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus'] = False

def imshow(name, img):
    temp = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.figure(figsize=(10,10))
    plt.title(name, fontsize='xx-large')
    plt.imshow(temp)    # 使用jupyter的时候好像没办法使用cv2.imshow，会卡
    
origin_image_path = r'C:\Users\ZerosZhang\Documents\PycharmCode\Learn_OpenCV_in_Python\image\pic_0010.jpg'
img = cv2.imread(origin_image_path, cv2.IMREAD_COLOR)
imshow('原始图片', img)
````

## 图像缩放 cv2.resize()
```python
dst = cv.resize(src, # 源图像
                dsize, # 输出图片的尺寸，如果给None，则使用比例因子fx和fy
                fx,
                fy, 
                interpolation=INTER_LINEAR)
```

`interpolation` 插值方法参数如下
- cv2.INTER_NEAREST:最近邻插值
- cv2.INTER_LINEAR:双线性插值，扩展时使用
- cv2.INTER_CUBIC:三线性插值，扩展时使用，比较慢
- cv2.INTER_AREA:使用像素面积关系进行重采样，缩放推荐

````python
dst_1 = cv2.resize(img, None, fx=2, fy=2, interpolation=cv2.INTER_CUBIC)
imshow('图像缩放', dst_1)  # 这里我由于已经定义好了plt窗口的大小，所以没有改变，可以使用cv2.imshow()查看
````

## 图像平移 cv2.warpAffine
$$
M = 
\begin{bmatrix} 
1 & 0 & t_x\\ 
0 & 1 & t_y\\ 
\end{bmatrix}
$$

左边四个表示旋转矩阵，当前为单位矩阵，右边两位表示平移量
```python
dst = cv.warpAffine(src, # 图像源
                    M, # 变换矩阵
                    dsize, # 输出图像的尺寸
                    dst, # 目标图像
                    flags=INTER_LINEAR, # 插值方法
                    borderMode=BORDER_CONSTANT, # 像素外推法
                    borderValue=0)
```

````python
translate = np.float32([[1,0,100],
                        [0,1,50]])
dst_2 = cv2.warpAffine(img, translate, img.shape[1::-1])
imshow('图像平移', dst_2)
````

## 图像旋转 cv2.getRotationMatrix2D()
图像旋转使用的就是上面的 `cv2.warpAffine()` 函数，但是旋转矩阵我们可以通过 `cv2.getRotationMatrix2D` 计算得到

```python
rotate = cv2.getRotationMatrix2D(center, # 旋转中心
                                angle, # 旋转角度
                                scale # 缩放因子)
```

````python
rows, cols = img.shape[:2]
rotate = cv2.getRotationMatrix2D((cols/2, rows/2), 45, 0.5)
dst_3 = cv2.warpAffine(img, rotate, img.shape[1::-1])
imshow('图像旋转缩放', dst_3)
````

## 仿射变换 cv2.getAffineTransform()

简单来说，“仿射变换” = “线性变换” + “平移”

![image-20210508095626683](https://cdn.jsdelivr.net/gh/sheng962464/PicGo/img/20210508095632.png)

上图中正方体的侧面就是一个仿射变换

仿射变换我们使用 `cv2.getAffineTransform()` 函数来获得仿射变换的矩阵
```python
affine = cv2.getAffineTransform(src, # src是三个点
                                dst # dst是三个点)
```

![image-20210508101844638](https://cdn.jsdelivr.net/gh/sheng962464/PicGo/img/20210508101845.png)

这里有一点需要注意的是，一开始我使用的是上图的坐标点进行计算，所获得的图像总是与我想要的图像相反
```python
src_point = np.float32([[0, 0], [3, 3], [3, -3]])
dst_point = np.float32([[0, 0], [6, 3], [0, -3]])
```
后来想到opencv的坐标系与笛卡尔坐标系是不一样的，因此实际应该使用下图所示的坐标点进行计算

![image-20210508103835134](https://cdn.jsdelivr.net/gh/sheng962464/PicGo/img/20210508103836.png)

````python
src_point = np.float32([[0, 0], [6, 0], [6, 6]])
dst_point = np.float32([[3, 0], [9, 0], [3, 6]]) 
affine = cv2.getAffineTransform(src_point, dst_point)
dst_4 = cv2.warpAffine(img, affine, img.shape[1::-1])
imshow('仿射变换', dst_4)
````

## 透视变换 cv2.getPerspectiveTransform()
透视变换简单来说就是“近大远小”

![poker](https://cdn.jsdelivr.net/gh/sheng962464/PicGo/img/20210508110848.jpg)

由于相机是斜着进行拍照片的，因此拍到的扑克牌有点倾斜，因此可以使用透视变换，来获得扑克牌的正面。

计算的方法和上面的仿射变换类似，提供4个对应的点对，通过 `cv2.getPerpectiveTransform()` 计算3*3透视矩阵，然后通过 `cv2.warpPerpective()` 对图片进行变换
```python
cv2.getPerspectiveTransform(src, # src是四个点
                            dst # dst是四个点  )
dst = cv.warpPerpective(src, # 图像源
                    M, # 变换矩阵
                    dsize, # 输出图像的尺寸
                    dst, # 目标图像
                    flags=INTER_LINEAR, # 插值方法
                    borderMode=BORDER_CONSTANT, # 像素外推法
                    borderValue=0)
```

````python
origin_image_path = r'C:\Users\ZerosZhang\Documents\PycharmCode\Learn_OpenCV_in_Python\image\pic_0011.jpg'
img = cv2.imread(origin_image_path, cv2.IMREAD_COLOR)
# 这里点是使用plt显示图片，然后手点坐标点出来的，不是很精确，但是够用了
src_point = np.float32([[208, 46],[397, 84],[138, 273],[351, 323]])
dst_point = np.float32([[0,0],[570,0],[0,880],[570,880]])
perpective = cv2.getPerspectiveTransform(src_point, dst_point)
dst_5 = cv2.warpPerspective(img, perpective, (570,880))
imshow('透视变换', dst_5)
````
