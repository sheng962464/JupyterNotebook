```python
import scipy
import numpy as np
import matplotlib.pyplot as plt
```

```python
point_path = r'd:\circle.txt'
point_list = []
with open(point_path, 'r', encoding='utf-8') as f:
    for line in f:
        x, y, z = map(float, line.split(','))
        point_list.append([x, y, z])
```

```python
x = [point[0] for point in point_list]
y = [point[1] for point in point_list]
```

```python
x_m = np.mean(x)
y_m = np.mean(y)

#圆心估计
center_estimate = np.array([x_m, y_m])
print(center_estimate)
```

```python
plt.plot(x,y,'o')
plt.title('Circle Fit') 
plt.axis('scaled')
plt.xlabel('X')
plt.ylabel('Y')
```

```python
x = y = np.arange(-4, 4, 0.1)
x, y = np.meshgrid(x,y)
plt.contour(x, y, x**2 + y**2, [9])     #x**2 + y**2 = 9 的圆形

plt.axis('scaled')
plt.show()
```

```python
x = 5  # 圆心的x轴坐标
y = 5  # 圆心的y轴坐标
r = 3  # 圆的半径
circle = plt.figure(figsize=(5,5))
circle = plt.Circle((x, y), r, color='y', fill=False)
plt.gcf().gca().add_artist(circle)
plt.axis('equal')
plt.xlim(0, 10)
plt.ylim(0, 10)
plt.show()
```
